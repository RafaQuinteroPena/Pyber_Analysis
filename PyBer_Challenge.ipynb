### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "city_data.csv"
ride_data_to_load = "ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
total_rides_per_city_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_per_city_type.head()
```

```python
# 2. Get the total drivers for each city type
total_drivers_per_city_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_per_city_type.head()
```

```python
#  3. Get the total amount of fares for each city type
total_fare_amount_per_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_amount_per_city_type.head()
```

```python
#  4. Get the average fare per ride for each city type. 
avg_fare_per_ride_by_city_type = total_fare_amount_per_city_type / total_rides_per_city_type
avg_fare_per_ride_by_city_type.head()
```

```python
# 5. Get the average fare per driver for each city type. 
avg_fare_per_driver_by_city_type = total_fare_amount_per_city_type / total_drivers_per_city_type
avg_fare_per_driver_by_city_type.head()
```

```python
#  6. Create a PyBer summary DataFrame. 
Pyber_Summary_df = pd.DataFrame({
    "Total Rides per City Type" : total_rides_per_city_type,
    "Total Drivers per City Type" : total_drivers_per_city_type,
    "Total Amount of Fare per City Type" : total_fare_amount_per_city_type,
    "Avg Fare per Ride by City Type" : avg_fare_per_ride_by_city_type,
    "Avg Fare per Driver by City Type" : avg_fare_per_driver_by_city_type})
Pyber_Summary_df.head()
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
Pyber_Summary_df = pd.DataFrame({
    "Total Rides per City Type" : total_rides_per_city_type,
    "Total Drivers per City Type" : total_drivers_per_city_type,
    "Total Amount of Fare per City Type" : total_fare_amount_per_city_type,
    "Avg Fare per Ride by City Type" : avg_fare_per_ride_by_city_type,
    "Avg Fare per Driver by City Type" : avg_fare_per_driver_by_city_type})

Pyber_Summary_df.index.name = None
Pyber_Summary_df.head()
```

```python
#  8. Format the columns.
Pyber_Summary_df = pd.DataFrame({
    "Total Rides" : total_rides_per_city_type,
    "Total Drivers" : total_drivers_per_city_type,
    "Total Fares" : total_fare_amount_per_city_type,
    "Average Fare per Ride" : avg_fare_per_ride_by_city_type,
    "Average Fare per Driver" : avg_fare_per_driver_by_city_type})

Pyber_Summary_df.index.name = None
Pyber_Summary_df.head()
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.

total_fare_per_date_type = pyber_data_df.groupby(["type","date"]).sum()[["fare"]]
total_fare_per_date_type

```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
# df = df.reset_index()
total_fare_per_date_type = total_fare_per_date_type.reset_index()
total_fare_per_date_type
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
total_fare_per_date_type_pivot = total_fare_per_date_type.pivot(index="date",columns="type",values="fare")
total_fare_per_date_type_pivot.head(10)
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
jan_2019_to_april_2019_df = total_fare_per_date_type_pivot.loc["2019-01-01":"2019-04-29"]
jan_2019_to_april_2019_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
# df.index = pd.to_datetime(df.index)
jan_2019_to_april_2019_df.index = pd.to_datetime(jan_2019_to_april_2019_df.index)
jan_2019_to_april_2019_df
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
jan_2019_to_april_2019_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
total_fare_per_city_type_weekly = jan_2019_to_april_2019_df.resample("W").sum()
total_fare_per_city_type_weekly.head(10)
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

total_fare_per_city_type_weekly.plot(figsize=(20 , 5))
plt.title("Total Weekly Fares per City Type", fontsize=20)
plt.ylabel("Fare", fontsize=18)
plt.xlabel("Month", fontsize=18)
plt.grid(True)

plt.legend()

plt.show()
```
